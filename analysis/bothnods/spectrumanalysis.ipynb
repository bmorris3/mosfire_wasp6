# Quickest Spectrum Analysis

Load spectrum, use MCMC to find best-fit linear "model spectrum", find out if the slope is consistent with zero or negative.

### Load spectrum

```python
%pylab inline
import pyfits
import numpy as np
import matplotlib
matplotlib.rcParams['font.size'] = 15
from matplotlib import pyplot as plt
import sys
sys.path.append('../')
import photPack2
from astropy.time import Time
import emcee

matplotlib.rcParams['font.size'] = 12


wavelengthbincenters = np.load('photoutputs/wavelengthbincenters.npy')
spectrummatrix = np.load('spectrum.npy')
spectrum = spectrummatrix[:,0]
spectrumpluserrors = spectrummatrix[:,1]
spectrumminuserrors = spectrummatrix[:,2]
print wavelengthbincenters,spectrum

fig, ax = plt.subplots(1, figsize=(16,12))

from scipy import optimize
linear = lambda x: x[0]*wavelengthbincenters + x[1]
errorfunc = lambda x: linear(x) - spectrum
bestp = optimize.leastsq(errorfunc, [0,0])[0]


ax.errorbar(wavelengthbincenters, spectrum, yerr=[spectrumminuserrors, spectrumpluserrors], \
              fmt='s', color='black', \
              ecolor='k', markeredgecolor='none', ms=10)
ax.plot(wavelengthbincenters, linear(bestp), lw=2, color='r', label='leastsq')

labelsize = 25
ax.set_xlabel('Wavelength [$\mu m$]', fontsize=labelsize)
ax.set_ylabel(r'$R_p/R_*$', fontsize=labelsize)
ax.legend()
#fig.savefig('/astro/users/bmmorris/Dropbox/2015aas225/plots/simplespectrum.eps', bbox_inches='tight')
plt.show()
```

### Fit Spectrum with MCMC

```python
def lnlike(theta, x, y, yerr):
    model = x*theta[0] + theta[1]
    return -0.5*(np.sum((y-model)**2/yerr**2))# - np.log(inv_sigma2)))

def lnprior(theta):
    slope, intercept = theta
    # If parameter is locked, limits are set to [0,0]. If parameter is not locked,
    # check that all values for that parameter are within the set limits. If they are,
    # return 0.0, else return -np.inf
    if not ((-1 < slope < 1) and (0 < intercept < 1)):
            return -np.inf
    return 0.0

def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)

Nfreeparameters = 2
ndim = Nfreeparameters
nwalkers = 2*Nfreeparameters if 2*Nfreeparameters % 2 == 0 else 2*Nfreeparameters + 1

meanerrs = np.mean([spectrumminuserrors, spectrumpluserrors], axis=0)
#pos = [initP + 1e-3*np.random.randn(len(initP)) for i in range(nwalkers)]
Nsteps = 5e4
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, \
          args=(wavelengthbincenters, spectrum, meanerrs))

print 'ndim =', ndim
print 'nwalkers =', nwalkers

print "Running initial burn in"
initP = bestp
pos = [initP + 1e-3*np.random.randn(len(initP)) for i in range(nwalkers)]
p0, _, _ = sampler.run_mcmc(pos, Nsteps)
samples = sampler.chain[:, int(0.1*Nsteps):, :].reshape((-1, ndim))
```

```python
import triangle
labels = ['slope', 'intercept']
trifig, ax = plt.subplots(Nfreeparameters, Nfreeparameters, figsize=(10, 10))
fig2 = triangle.corner(samples[:, :], labels=labels, 
                       fig=trifig, plot_datapoints=False) # truths=[t0_expected, aOverRs, RpOverRs, scale, 1]
plt.show()
```

### How flat is the spectrum?

```python
slopesamples = samples[:,0]
intsamples = samples[:,1]
def medplusminus(vector):
    v = np.percentile(vector, [16, 50, 84])
    return v[1], v[2]-v[1], v[1]-v[0]
sloperesult = medplusminus(slopesamples)
intresult = medplusminus(intsamples)

print (sloperesult[0] - 0)/np.mean(sloperesult[1:]),'sigma from flat (slope=0)'

fig, ax = plt.subplots(1, figsize=(16,12))

ax.errorbar(wavelengthbincenters, spectrum, yerr=[spectrumminuserrors, spectrumpluserrors], \
              fmt='s', color='black', \
              ecolor='k', markeredgecolor='none', ms=10)

for s in samples[np.random.randint(len(samples), size=1000)]:
    ax.plot(wavelengthbincenters, linear(s), lw=2, color='k', alpha=0.02)

mcmc_bestp = [sloperesult[0], intresult[0]]
ax.plot(wavelengthbincenters, linear(mcmc_bestp), lw=3, color='r', label='Max $\log\, p$')
    
labelsize = 25
ax.set_xlabel('Wavelength [$\mu m$]', fontsize=labelsize)
ax.set_ylabel(r'$R_p/R_*$', fontsize=labelsize)
ax.legend()
#fig.savefig('/astro/users/bmmorris/Dropbox/2015aas225/plots/simplespectrum.eps', bbox_inches='tight')
plt.show()
```
