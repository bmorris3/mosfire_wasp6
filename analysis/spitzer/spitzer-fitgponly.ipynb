```python
import sys
sys.path.append('/astro/users/bmmorris/Downloads/Fast_MA')
from ext_func.rsky import rsky
from ext_func.occultquad import occultquad

def get_lc(aRs, i, t0, q1, q2, p0, F0, e, w, period, eps, t):
    '''
    e - eccentricity
    aRs - "a over R-star"
    i - inclination angle in radians
    u1, u2 - quadratic limb-darkening coeffs
    p0 - planet to star radius ratio
    w - argument of periapse
    period - period
    t0 - midtransit (JD)
    eps - minimum eccentricity for Kepler's equation
    t - time array
    '''
    u1 = 2*np.sqrt(q1)*q2
    u2 = np.sqrt(q1)*(1 - 2*q2)
    r_s = 1.0
    npoints = len(t)
    z0 = rsky(e, aRs, i, r_s, w, period, t0, eps, t)   #calculates separation of centers between the planet and the star
    mu_c = occultquad(z0, u1, u2, p0, npoints)   #returns limb darkened model lightcurve
    return F0*mu_c

from scipy import optimize
aOverRs = 1./0.0932 # Jord`an et al 2013
RpOverRs = 0.1404   # Jord`an et al 2013
eccentricity = 0.054
inclination = np.radians(88.47)
q1 = 0.00001
q2 = 0.2
periapse = np.pi/2
period = 3.361006
mineccentricity = 1.0e-7
t0_roughfits = [2456313.94851, 2456307.23216]

paramlimits = 2*[[-50, 50]] + 2*[[-50, 50]] + 2*[[-12, -10]]
print paramlimits
               #[-40, 1], [-40, 1]]
genmodel1 = lambda p: np.ones_like(ch1['t'])
genmodel2 = lambda p: np.ones_like(ch2['t'])

# initP = [-11, -11]
initP = 2*[-13] + 2*[-14] + 2*[-10.85]
print initP
expinitP = np.exp(initP)
kernel1 = expinitP[0]*kernels.Matern32Kernel(expinitP[2]) + kernels.WhiteKernel(expinitP[4])
kernel2 = expinitP[1]*kernels.Matern32Kernel(expinitP[3]) + kernels.WhiteKernel(expinitP[5])
# kernel1 = expinitP[0]*kernels.ExpKernel(expinitP[2]) + kernels.WhiteKernel(expinitP[4])
# kernel2 = expinitP[1]*kernels.ExpKernel(expinitP[3]) + kernels.WhiteKernel(expinitP[5])
# kernel1 = kernels.WhiteKernel(expinitP[0])
# kernel2 = kernels.WhiteKernel(expinitP[1])
gp1 = george.GP(kernel1)
gp2 = george.GP(kernel2)
gp1.compute(ch1['t'])#, ch1['e'])
gp2.compute(ch2['t'])#, ch2['e'])

def lnlike(theta):
#     c1, c2 = np.exp(theta[:2])
    a1, a2, tau1, tau2, c1, c2 = np.exp(theta[:6])
    model1 = genmodel1(theta)
    model2 = genmodel2(theta)

#     gp1 = george.GP(a1 * kernels.Matern32Kernel(tau1) + kernels.WhiteKernel(c1))#, solver=george.HODLRSolver)
#     gp1 = george.GP(kernels.WhiteKernel(c1))#, solver=george.HODLRSolver)
#     gp1.compute(ch1['t'], ch1['e'])
    
#     gp2 = george.GP(a2 * kernels.Matern32Kernel(tau2) + kernels.WhiteKernel(c2))#, solver=george.HODLRSolver)
#     gp2 = george.GP(kernels.WhiteKernel(c2))#, solver=george.HODLRSolver)
#     gp2.compute(ch2['t'], ch2['e'])

    gp1.kernel.pars = [a1, tau1, c1]
    gp2.kernel.pars = [a2, tau2, c2]
#     gp1.kernel.pars = [c1]
#     gp2.kernel.pars = [c2]
    return gp1.lnlikelihood(ch1['f'] - model1) + gp2.lnlikelihood(ch2['f'] - model2)


def lnprior(theta, paramlimits=paramlimits):
    # If parameter is locked, limits are set to [0,0]. If parameter is not locked,
    # check that all values for that parameter are within the set limits. If they are,
    # return 0.0, else return -np.inf
    for i, limits in enumerate(paramlimits):
        if not ((limits[0] < theta[i]) and (theta[i] < limits[1])):
            #print limits, parameters[i]
            return -np.inf
    return 0.0

def lnprob(theta):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta) if np.isfinite(lp) else -np.inf
```

```python
def gelmanrubin(samples, useparameter):
    # W = mean of within-chain variance
    # B = between chain variance
    n, m = np.shape(samples)
    individualchains = [samples[i:, useparameter][::2*m] for i in range(2*m)]
    W = np.mean([np.var(chain, ddof=1) for chain in individualchains])
    B = n*np.var([np.mean(chain) for chain in individualchains], ddof=1)

    Vhat = W*(n-1)/n + B/n
    Rhat = np.sqrt(Vhat/W)
    return Rhat

fig, ax = plt.subplots(1, figsize=(16,5))
for i in range(len(labels)):
    ax.bar(i-0.5, gelmanrubin(samples, i), color='k')
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, ha='right')
[l.set_rotation(45) for l in ax.get_xticklabels()]
ax.set_ylim([1,2])
ax.set_xlim([-1, len(labels)+1])
ax.set_ylabel('$\hat{R}$')
ax.set_title('Gelman-Rubin Statistic')
plt.show()
```
