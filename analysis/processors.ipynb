```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

workstations = 1600 # MHz
magneto = 1200 # MHz

Ncores = np.arange(4,15)

workstation_cycles = 4.*workstations
magneto_cycles = Ncores*magneto

fig, ax = plt.subplots(1, 2, figsize=(10,6))
ax[0].plot(Ncores, magneto_cycles, 'o-', 
           label='Magneto: {0:d}MHz'.format(magneto))
ax[0].plot(Ncores, workstation_cycles*np.ones_like(Ncores), 'o-', 
           label='Local: {0:d}MHz'.format(workstations))
ax[0].set_xlabel('cores')
ax[0].set_ylabel('MHz')
ax[0].set_title('Speed')
ax[0].legend(loc='upper left')

ax[1].plot(Ncores, magneto_cycles/workstation_cycles, 'o-', label='Local')
ax[1].set_xlabel('cores')
ax[1].set_ylabel('Factor improvement')
ax[1].set_title('Improvement using magneto')
ax[1].axhline(1, c='k', ls='--')

#ax[0].annotate('Workstation: {0:d}MHz\nMagneto: {1:d}MHz'.format(
#        workstations, magneto), (0.1, 0.8), xycoords='axes fraction')

fig.savefig('plots/magneto.pdf',bbox_inches='tight')
plt.show()
```
