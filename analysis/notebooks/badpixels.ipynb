```python
%pylab inline
import pyfits
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.append('../')
import photPack2
from astropy.time import Time

masterflatdark = pyfits.getdata('/local/tmp/mosfire/2014sep18_analysis/masterflatdark.fits')
badpixpath = '/astro/store/scratch/tmp/bmmorris/mosfire/2014sep18_analysis/badpix_10sep2012.fits'
badpixelmask = pyfits.getdata(badpixpath)

n, edges = np.histogram(masterflatdark, 100)
bincenters = [0.5*(edges[i-1]+edges[i]) for i in range(1,len(edges))]
s = np.std(masterflatdark)
m = np.median(masterflatdark)
thres = 2.5
fig, ax = plt.subplots(1,3,figsize=(17,7))
ax[0].semilogy(bincenters, n, 'k')
ax[0].axvline(m + thres*s)
ax[0].axvline(m - thres*s)

n, edges = np.histogram(masterflatdark, 100, range=[m-thres*s,m+thres*s])
bincenters = [0.5*(edges[i-1]+edges[i]) for i in range(1,len(edges))]
ax[1].semilogy(bincenters,n,'k')
ax[1].axvline(m + thres*s)
ax[1].axvline(m - thres*s)

badpixels_flat = (masterflatdark > m + thres*s)+(masterflatdark < m - thres*s)
badpixelimage_flat = masterflatdark
badpixelimage_flat[np.invert(badpixels_flat)] = 0
badpixelimage_flat[badpixels_flat] = 1
ax[2].imshow(badpixelimage_flat, interpolation='nearest',origin='lower',cmap='binary')
plt.show()
print sum(badpixels_flat)
print sum(badpixelmask)
```

```python
rowlimits = [5, 2030]
collimits = [5, 2044]
sampleimage = pyfits.getdata('/local/tmp/mosfire/2014sep18_analysis/m140918_0627n.fits')
badpixelmask = pyfits.getdata(badpixpath)[rowlimits[0]:rowlimits[1],collimits[0]:collimits[1]]
bestxshifts = np.load('/local/tmp/mosfire/2014sep18_analysis/bestxshifts.npy')
def channelshift(image):
    oversamplefactor = 1
    ydim, xdim = image.shape
    outputpaddingwidth = np.ceil(np.max(bestxshifts)/oversamplefactor)
    outputpadding = np.zeros((ydim, outputpaddingwidth))
    paddedimage = np.hstack([outputpadding, image, outputpadding])

    for row in range(1, ydim):
        paddedimage[row] = np.roll(paddedimage[row], int(bestxshifts[row]/oversamplefactor))
    return paddedimage

shiftedbadpixelmask = channelshift(badpixelmask)
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(shiftedbadpixelmask, origin='lower')
plt.show()
pyfits.writeto('/local/tmp/mosfire/2014sep18_analysis/badpix_10sep2012_shifted.fits',shiftedbadpixelmask)
```
