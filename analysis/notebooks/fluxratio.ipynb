# Flux ratio between bright and dim source

There's a dim source right next to the target star. It's too dim even to distinguish if it's a point source or not, but it will be a small source of contamination in the WASP-6 aperture that we should account for. To do that accounting, I've tried to estimate the dim source's brightness below. I checked with Vizier that there are no known sources in that position with that brightness, that I could use to double check my brightness measurement.
