# NICFPS Imaging

Using imaging from [NICFPS](http://www.apo.nmsu.edu/arc35m/Instruments/NICFPS/) on the [ARC 3.5 meter](http://www.apo.nmsu.edu/arc35m/) at Apache Point Observatory on WASP-6 to identify and get a photometric measurement on the contaminating source in our nod subtraction in the MKO $K_s$ band.

## Make flat field

```python
%pylab inline
import pyfits
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.append('../')
import photPack2
from astropy.time import Time
from scipy import optimize
from scipy.ndimage import gaussian_filter1d, gaussian_filter
from glob import glob

skyflatpaths = glob('/astro/store/scratch/tmp/bmmorris/APO/UT141110/skyflat*')
flatdarkpaths = glob('/astro/store/scratch/tmp/bmmorris/APO/UT141110/NICFPSdark-2s*')

# Make dark frame for flats
flatdark = np.zeros((1024, 1024, len(flatdarkpaths)), dtype=np.float64)
for i, path in enumerate(flatdarkpaths):
    flatdark[:,:,i] = pyfits.getdata(path)
flatdark = np.median(flatdark, axis=2)

# Make sky flat by median combining five dither positions
skyflat = np.zeros((1024, 1024, len(skyflatpaths)), dtype=np.float64)
for i, path in enumerate(skyflatpaths):
    skyflat[:,:,i] = pyfits.getdata(path) - flatdark
skyflat = np.median(skyflat, axis=2)
skyflat /= np.median(skyflat)
pyfits.writeto('tmp/skyflat.fits', skyflat, clobber=True)

plotflat = False
if plotflat:
    fig, ax = plt.subplots(figsize=(12,12))
    ax.imshow(skyflat, origin='lower', interpolation='nearest', cmap='Blues_r')
    plt.show()
```

## WASP-6 Nod-subtracted, stacked image

```python

firstind = 366
lastind = 519
evenpaths = ['/astro/store/scratch/tmp/bmmorris/APO/UT141110/WASP-6dither.%04d.fits' % i \
            for i in range(firstind, lastind-1, 2)]
oddpaths = ['/astro/store/scratch/tmp/bmmorris/APO/UT141110/WASP-6dither.%04d.fits' % i \
            for i in range(firstind+1, lastind, 2)]

diffimg = np.zeros((1024, 1024), dtype=np.float64)
sumimg = np.zeros((1024, 1024), dtype=np.float64)
for evenpath, oddpath in zip(evenpaths, oddpaths):
    e = pyfits.getdata(evenpath)
    o = pyfits.getdata(oddpath)
    sumimg += e/skyflat + o/skyflat
    diffimg += e/skyflat - o/skyflat
```

### Do photometry on the dim source and on WASP-6

```python
sys.path.append('radialphotometrytools')
import photometry
import trackSmooth
reload(photometry)
def trackandphot(imagedata, imagesum, init_x, init_y, plottingThings):
    smoothconst = 3.
    zoomfactor = 25
    trackplots = True
    apertureradius = 25
    [xCenter,yCenter,averageRadius, _] = trackSmooth.trackSmooth(imagedata,init_y, init_x, \
        smoothconst, plottingThings, preCropped=False, zoom=zoomfactor, plots=trackplots)
    flux, error, photFlags, _ = photometry.phot_sumbg(imagedata, imagesum, xCenter, yCenter,
                                                    apertureradius,
                                                    plottingThings,
                                                    ccdGain=1.0,
                                                    plots=trackplots,\
            annulusOuterRadiusFactor=2.5, annulusInnerRadiusFactor=1.1,\
            sigmaclipping=True, returnsubtractedflux=True)
    plt.show()
    return xCenter, yCenter, flux, error

# Set up plot
fig = plt.figure(num=None, figsize=(14, 3), facecolor='w', edgecolor='k')
fig.subplots_adjust(wspace=0.5)
subplotsDimensions = 140
photSubplotsOffset = 3
statusSubplotOffset = 6
plottingThings = [fig,subplotsDimensions,photSubplotsOffset]

dimcentroid_init = [487.56248, 609.70507]#[::-1]
brightcentroid_init = [512.19206, 481.9989]#[::-1]
dimcentroid_x, dimcentroid_y, dimflux, dimerr  = trackandphot(diffimg, sumimg, dimcentroid_init[0], \
                                                              dimcentroid_init[1], plottingThings)
# Set up second plot
fig = plt.figure(num=None, figsize=(14, 3), facecolor='w', edgecolor='k')
fig.subplots_adjust(wspace=0.5)
subplotsDimensions = 140
photSubplotsOffset = 3
statusSubplotOffset = 6
plottingThings = [fig,subplotsDimensions,photSubplotsOffset]
brightcentroid_x, brightcentroid_y, brightflux, brighterr  = trackandphot(diffimg, sumimg, brightcentroid_init[0], \
                                                              brightcentroid_init[1], plottingThings)
plt.show()
```

```python
fluxratio = dimflux/brightflux
fluxratio_unc = fluxratio*np.sqrt( (dimerr/dimflux)**2 + (brighterr/brightflux)**2 )
magdiff = -2.5*np.log10(fluxratio)
magdiff_unc = np.sqrt((-2.5/fluxratio/np.log(10)*fluxratio_unc)**2)

K2MASS_WASP6mag = 10.325 # 2MASS Ks band
K2MASS_WASP6mag_err = 0.025 # 2MASS Ks band

print 'Flux ratio = %f +/- %f' % (fluxratio, fluxratio_unc)
print 'Magnitude difference = %f +/- %f' % (magdiff, magdiff_unc)
print 'Magnitude of WASP-6 = K_MKO = %f' % K2MASS_WASP6mag
print 'Flux of dim source = **THIS IS NOT YET CORRECTED FOR DIFFERENCE IN FILTER SETS** = %f' % (K2MASS_WASP6mag+magdiff)
```

*Notes on uncertainties*: If $f(x) = \log_b(x)$ then $f'(x) = \frac{1}{b\ln x}$ so 

$$ \sigma_\text{(magdiff)} = \sqrt{ \left( \frac{\partial \text{(magdiff)}}{\partial \text{(fluxratio)}} \sigma_\text{(fluxratio)} \right)^2 } $$
$$ \sigma_\text{(magdiff)} = \sqrt{ \left( \frac{-2.5}{\text{(fluxratio)} \ln 10} \sigma_\text{(fluxratio)} \right)^2 } $$

*Notes on bandpass conversion*: Using the conversion from 2MASS K band to MKO Ks band from the [2MASS webpage](http://www.astro.caltech.edu/~jmc/2mass/v3/transformations/) --
$$ (K_s)_{2MASS} = K_{MKO} + (0.002 \pm 0.004) + (0.026 \pm 0.006)(J-K)_{MKO}$$
