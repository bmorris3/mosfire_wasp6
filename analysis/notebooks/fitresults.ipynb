```python
import sys
sys.path.append('/astro/users/bmmorris/Downloads/Fast_MA')
from ext_func.rsky import rsky
from ext_func.occultquad import occultquad
def get_lc(e, aRs, i, q1, q2, p0, w, period, t0, eps, t):
    '''
    e - eccentricity
    aRs - "a over R-star"
    i - inclination angle in radians
    u1, u2 - quadratic limb-darkening coeffs
    p0 - planet to star radius ratio
    w - argument of periapse
    period - period
    t0 - midtransit (JD)
    eps - minimum eccentricity for Kepler's equation
    t - time array
    '''
    u1 = 2*np.sqrt(q1)*q2
    u2 = np.sqrt(q1)*(1 - 2*q2)
    r_s = 1.0
    npoints = len(t)
    z0 = rsky(e, aRs, i, r_s, w, period, t0, eps, t)   #calculates separation of centers between the planet and the star
    mu_c = occultquad(z0, u1, u2, p0, npoints)   #returns limb darkened model lightcurve
    return mu_c
    
#from scipy import optimize
aOverRs = 1./0.0932 # Jord`an et al 2013
RpOverRs = 0.1404   # Jord`an et al 2013
eccentricity = 0
inclination = 88.47*np.pi/180
#u1 = 0.2
#u2 = 0.2
q1 = 0.2
q2 = 0.2
periapse = np.pi/2
period = 3.361006
mineccentricity = 1.0e-7
times_JD = times

def fitfunc_bin(p):
    #return get_lc(eccentricity, aOverRs, inclination, u1, u2, RpOverRs, periapse, period, t0, mineccentricity, times)
    return p[4]*get_lc(eccentricity, aOverRs_bandint, inclination, p[0], p[1], p[2], \
           periapse, period, t0_bandint, mineccentricity, times) + p[3]*airmass
```

```python
from glob import glob

mcmcpickles = glob('outputs/bins*.npy')
times = np.load('times.npy')
wavelengthsoln = np.load('wavelengthsoln.npy')
bandintegratedsamples = np.load('bandintegratedchain.npy')
# aOverRs_mcmc, q1_mcmc, q2_mcmc, RpOverRs_mcmc, t_mcmc, airmassconst_mcmc, scale_mcmc, f_mcmc = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]),
#                              zip(*np.percentile(bandintegratedsamples, [16, 50, 84],
#                             axis=0)))
aOverRs_bandint, q1_bandint, q2_bandint, RpOverRs_bandint, t0_bandint, airmassconst_bandint, scale_bandint, f_bandint  = np.median(bandintegratedsamples, axis=0)

def chi2(v1, v2, err, Nfreeparams):
    return np.sum( ((v1-v2)/err)**2 )/(len(v1) - Nfreeparams)

#fig, ax = plt.subplots(2,10,figsize=(16,14), sharex=True)
mininttime = int(np.min(times))
def onebinlc(ax, binindex, offset=0):
    # Next add chi^2 annotation
    lc = fluxes[:, 1, binindex]/fluxes[:, 0, binindex]
    lc_errors = lc*np.sqrt((errors[:, 1, binindex]/fluxes[:, 1, binindex])**2 +\
-                (errors[:, 0, binindex]/fluxes[:, 0, binindex])**2)
    samples = np.load('outputs/bin%02d.npy' % binindex)
    q1_mcmc, q2_mcmc, RpOverRs_mcmc, airmassconst_mcmc, scale_mcmc, f_mcmc = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]),
                                 zip(*np.percentile(samples, [16, 50, 84], axis=0)))
    bestfitparams = [i[0] for i in [q1_mcmc, q2_mcmc, RpOverRs_mcmc, airmassconst_mcmc, scale_mcmc, f_mcmc]]
    model = fitfunc_bin(bestfitparams)
    
    ax[0, binindex-offset].errorbar(times - mininttime, lc, yerr=lc_errors, fmt='.', color='k')
    ax[0, binindex-offset].set_title('%.3f$\mu m$' % wavelengthbincenters[binindex], fontsize=18)
    ax[0, binindex-offset].plot(times - mininttime, model, 'r', lw=2)

    chisquared = chi2(lc, model, lc_errors, len(bestfitparams))
    text = r'$\chi^2 =' + ('%.3f$' % chisquared)
    ax[0, binindex-offset].annotate(text, xy=(0.5, 0.92), \
                                    xycoords='axes fraction', textcoords='axes fraction', \
                                    va='center', ha='center', fontsize=18)
    
    ax[1, binindex-offset].errorbar(times - mininttime, lc-model, yerr=lc_errors, fmt='.', color='k')
    ax[1, binindex-offset].axhline(0, color='k', ls='--', lw=1)

print 2*'\n'
fig, ax = plt.subplots(2,5,figsize=(18,14), sharex=True, sharey='row')
for i in range(5):#[0, 2, 3, 4]:#
    onebinlc(ax, i)
subplotskwargs = {'hspace' : 0, 'wspace' : 0.0 }
plt.subplots_adjust(**subplotskwargs)
fig.savefig('plots/results_0.pdf')
plt.show()

fig, ax = plt.subplots(2,5,figsize=(18,14), sharex=True, sharey='row')
for i in range(5,10):
    onebinlc(ax, i, offset=5)
plt.subplots_adjust(**subplotskwargs)
fig.savefig('plots/results_1.pdf')
plt.show()
```
